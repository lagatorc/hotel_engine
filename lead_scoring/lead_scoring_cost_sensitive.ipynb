# Data
- Lead scoring dataset from UCI
    - https://archive.ics.uci.edu/dataset/222/bank+marketing

# Methods
- Feature Engineering
    - One hot encoded categorical features
    - Scaled numerical features to have mean of 0 and variance of 1
- Model
    - Logistic regression with class weights to account for data imbalance
    - Grid search and cross validation for choosing parameters
    - L1 regularization used for feature selection
    - Platt scaling was used to calibrate the output probabilities
        - Makes probabilities more interpretable, predicted probability of 0.80 roughly represents an 80% chance of lead being converted after calibration.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.calibration import calibration_curve, CalibrationDisplay, CalibratedClassifierCV

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/raw/bank-additional/bank-additional-full.csv", delimiter=";")
```

```python
df.dtypes
```

```python
df = df.drop(columns=["duration", "previous", "poutcome"])
```

```python
cat_cols = ["job", "marital", "education", "default", 
            "housing", "loan", "contact", "month", 
            "day_of_week", "poutcome"]
num_cols = df.columns.difference(cat_cols + ["y"]).tolist()
for c in cat_cols:
    print(f"{c}: {len(df[c].unique())} categories")
```

```python
df[num_cols].describe()
```

```python
for n in num_cols:
    plt.figure()
    sns.kdeplot(df, x=n, hue="y")
```

```python
for n in cat_cols:
    plt.figure()
    plt.xticks(rotation=90)
    sns.histplot(df, x=n, hue="y", multiple="dodge")
```

```python
ohe = OneHotEncoder(sparse_output=False)
ohe_features = ohe.fit_transform(df[cat_cols])
ohe_cat_features = pd.DataFrame(ohe_features, columns=ohe.get_feature_names_out().tolist())
data = pd.concat([df[num_cols], ohe_cat_features], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, df["y"], test_size=0.3)
ct = ColumnTransformer([
        ('scaler', StandardScaler(), num_cols)
    ], remainder='passthrough')

x_train_scaled = ct.fit_transform(x_train)
x_test_scaled = ct.transform(x_test)
```

```python
params = {
    "penalty": ["l1"],
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "solver": ["saga"],
    "class_weight": ["balanced"],
    "max_iter": [500]
}
lr = LogisticRegression()
model = GridSearchCV(lr, params, scoring='roc_auc')
model.fit(x_train_scaled, y_train)
```

```python
model.best_params_
```

```python
feature_coeff = model.best_estimator_.coef_
feature_weights = pd.DataFrame({
    "feature": x_train.columns,
    "weight": np.sort(feature_coeff)[0][::-1]
})
ax = sns.barplot(
    feature_weights[np.abs(feature_weights["weight"]) > 1e-6],
    y = "feature",
    x = "weight",
    orient="h"
)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=5)
plt.tight_layout()
plt.show()
```

```python
y_pred = model.best_estimator_.predict_proba(x_test_scaled)
roc_auc_score(y_test, y_pred[:,-1])
```

```python
fpr, tpr, score = roc_curve(y_test, y_pred[:,-1], pos_label="yes")
sns.lineplot(x=fpr, y=tpr)
```

```python
sns.histplot(y_pred[:, 1])
```

```python
CalibrationDisplay.from_estimator(model.best_estimator_, x_test_scaled, y_test)
```

```python
calibrated_model = CalibratedClassifierCV(model.best_estimator_, cv="prefit")
calibrated_model.fit(x_test_scaled, y_test)
```

```python
sns.histplot(calibrated_model.predict_proba(x_test_scaled)[:, -1])
```

```python
CalibrationDisplay.from_estimator(calibrated_model, x_test_scaled, y_test)
```

```python
y_pred_calibrated = calibrated_model.predict_proba(x_test_scaled)
roc_auc_score(y_test, y_pred_calibrated[:,-1])
```

```python
fpr, tpr, score = roc_curve(y_test, y_pred_calibrated[:,-1], pos_label="yes")
sns.lineplot(x=fpr, y=tpr)
```
